### Importações

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Caminho do arquivo
arquivo = '/content/drive/MyDrive/Colab Notebooks/pandas/formulario_fofoqueiro.csv'
```

**Dados do arquivo**

```python
afrodev = open(arquivo)

csvreader = csv.reader(afrodev, delimiter=';')

header = []
header = next(csvreader)

rows = []
for row in csvreader:
  rows.append(row)

alunos_afrodev = np.array(rows)
afrodev.close()
```

## **Tratamento dos dados**

### Salário dos sonhos

```python
salarios_sonho = alunos_afrodev[:,9]
def tratamento_dados(valor):
  dado_tratado = valor.replace('R$', '').replace('k', '000').replace(' ', '').replace('mil', '000').replace('+', '').replace('mensais', '').replace(',00', '').replace('reais','')
  dado_tratado = dado_tratado.replace('.', '')
  return dado_tratado
salarios = np.array([float(tratamento_dados(i)) for i in salarios_sonho])
salarios
```

### Idade

```python
idades = np.array([int(s.split(' ')[0]) for s in alunos_afrodev[:,3]])
```

### CEP

```python
{len(x) for x in alunos_afrodev[:, 4]}
```

```python
def tratamento_cep(valor):
  if len(valor) == 7:
    valor = '0' + valor
  if len(valor) == 8:
    dado_tratado = valor
  else:
    dado_tratado = valor.replace('-', '')
  return dado_tratado
ceps = np.array([tratamento_cep(i) for i in alunos_afrodev[:, 4]])
ceps
```

### **Dicionário**

```python
data = {}
dados = []
for i in range(len(header)):
  for j in range(len(alunos_afrodev)):
    dados.append(alunos_afrodev[j][i])
  data[header[i]] = dados
  dados = []
```

```python
{x for x in data}
```

### **Dados tratados**

```python
# del data['Carimbo de data/hora']
data['Qual seu salário dos sonhos ? '] = salarios
data['Idade'] = idades
data['CEP'] = ceps
```

## **Dataframe**





```python
df = pd.DataFrame(data)
display(df)
```

### **Exclusão das colunas desnecessárias**

**`.drop(x, y)`**
- Remove linha ou coluna definida
- x => nome da coluna ou código da linha a ser removida
- y => qual dos eixos será excluido
  - `0 ou 'index'` => linha
  - `1 ou 'columns'` => coluna

```python
df = df.drop(['Carimbo de data/hora'], axis=1)
df = df.drop(['User_id'], axis=1)
```

### Tratando o tipo do dado

**`pd.to_numeric(x, y)`**
- Transforma a coluna em `int` ou `float`
- x => coluna que deverá ser transformada
- y => definição do que será feito em caso de erro

```python
df.dtypes
```

```python
df['O quão atraente você se considera de 0 até 10'] = pd.to_numeric(df['O quão atraente você se considera de 0 até 10'])
```
