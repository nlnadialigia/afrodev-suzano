## Séries

- objeto do tipo array unidimensional contendo uma sequência de valores e um array associado de rótulos(labels) de dados, chamado de índice(index)
- caso o índice não seja informado, será numerado automaticamente
- por meio dos atributos values e index, podemos obter esses valores da série
- os valores podem ser acessado como nos vetores

```python
obj = pd.Series([4, 7, -5, 3])
```

```python
print(obj.values)
```

```python
print(obj.index)
```

```python
print(obj)
```

```python
obj2 = pd.Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])
```

```python
print(obj2)
```

```python
obj[0]
```

```python
obj2['d']
```

```python
obj2[['a', 'b', 'c']]
```

```python
obj2['d'] = 6
```

```python
obj2['d']
```

```python
# busca na série

obj2[obj2 > 0]
# imprimir todos os números maiores que 0
```

```python
# Operações aritimeticas são todos os elmentos do array

obj2 *2
```

```python
np.abs(obj2)
```

```python
# Série como um dicionário ordenado de tamanho fixo, como se fosse um 
# mapeamento entre valores de índices e valores de dados.

'b' in obj2 # Validando se existe o indice
```

```python
# Criar uma série como um dicionário

sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj3 = pd.Series(sdata)
obj3
```

```python
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj4 = pd.Series(sdata, index=states)
obj4
```

```python
obj4.values
```

```python
pd.isnull(obj4)
```

```python
pd.notnull(obj4)
```

```python
obj4.isnull()
```

```python
# só soma os índices que tem nos dois objetos => retorna NaN para o valores que
# estão em um único objeto

print(f'obj3 =>\n{obj3}')
print(f'obj4 =>\n{obj4}')
print(f'soma =>\n{obj3 + obj4}')
```
